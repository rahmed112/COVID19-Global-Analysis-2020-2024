## l) **Yearly Analysis of Total Deaths, Total Cases, and Total Vaccinations** (Bar Graph)

This multivariate analysis will provide us a comprehensive overview of the total number of COVID-19 cases, deaths, and vaccinations from 2020 to 2024.

```python
# Select specific years
selected_years = [2020, 2021, 2022, 2023, 2024]
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
filtered_data = df[df['Year'].isin(selected_years)]

# Group data by year and sum total deaths, total cases, and total vaccinations
yearly_data = filtered_data.groupby('Year')[['Total deaths', 'Total cases', 'Total vaccinations']].mean().reset_index()

# Convert large numbers to millions for better readability
yearly_data['Total deaths'] = yearly_data['Total deaths'] / 1e6
yearly_data['Total cases'] = yearly_data['Total cases'] / 1e6
yearly_data['Total vaccinations'] = yearly_data['Total vaccinations'] / 1e6

# Define positions and width for bars
bar_width = 0.35
x = range(len(yearly_data['Year']))

# Create the figure and axis
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot total cases and total vaccinations on primary y-axis
bars1 = ax1.bar(x, yearly_data['Total cases'], width=bar_width, label='Total Cases (in millions)', color='orange')
bars2 = ax1.bar([p + bar_width for p in x], yearly_data['Total vaccinations'], width=bar_width, label='Total Vaccinations (in millions)', color='green')

# Customization for primary y-axis
ax1.set_xlabel('Year', fontsize=14, color='red')
ax1.set_ylabel('Count (in millions)', fontsize=14, color='red')
ax1.set_title('Yearly Analysis of Total Deaths, Total Cases, and Total Vaccinations', fontsize=16, fontweight='bold')
ax1.set_xticks([p + bar_width/2 for p in x])
ax1.set_xticklabels(yearly_data['Year'], rotation=45)
ax1.legend(loc='upper left')
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Create a secondary y-axis for total deaths
ax2 = ax1.twinx()
ax2.plot([p + bar_width/2 for p in x], yearly_data['Total deaths'], color='blue', marker='o', linestyle='-', linewidth=2, label='Total Deaths (in millions)')
ax2.set_ylabel('Total Deaths (in millions)', fontsize=14, color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Customization for secondary y-axis
ax2.legend(loc='center left')

# Adding data labels for total deaths
for i, txt in enumerate(yearly_data['Total deaths']):
    ax2.text(i + bar_width/2, txt, f'{txt:.2f}M', ha='right', va='bottom', fontsize=10, color='red')

# Adding data labels for total cases and total vaccinations
for bar in bars1:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}M', ha='center', va='bottom', fontsize=10, color='black')

for bar in bars2:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_y() + yval, f'{yval:.2f}M', ha='center', va='bottom', fontsize=10, color='black')

# Show the plot
plt.tight_layout()
plt.show()
```
